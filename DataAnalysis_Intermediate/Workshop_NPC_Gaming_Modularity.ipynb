```python
import random
```

```python
def success():
    rand = random.randrange(0,100) 
    if rand > 50:
        return True
    else:
        return False
```

```python
class Soul(object):
    def __init__(self,name,hp,mana):
        self.name=name
        self.hp = hp
        self.mana = mana
     
    
class Human(object):
    def __init__(self,power,target,name,hp,mana):
        #super().__init__(name,hp,mana)
        self.power=power
        self.target=target
        self.name=name
        self.hp = hp
        self.mana = mana
        print(f'My power is an {self.power}')
        print(f'My target is the {self.target}') 
        
    def attack(self, obj):
            inicio =1000
            fullMana=200
            
            if self.mana == 0 or self.hp == 0:
                print("You loose one turn to recover your mana and health")
            else:
                if success():
                    obj.hp == inicio
                    print("I defeated the monster and I have full health points")
                    obj.mana = fullMana - 10
                    print(f'I\'ve used 10 mana points but it was worth it')
                    print("My new mana points are:"), print(obj.mana)
                else:
                    obj.hp == inicio - 100
                    print("I have lost a battle but not the war")
                    obj.mana = fullMana - 100
                    print("I lost 100 mana, I must win next battle or I will have to retrieve")
                    print("I will change my weapon")
                    print("My new mana points are:"), print(obj.mana)

                    
class Monster(object):
    def __init__(self,name,hp,mana):
        #super().__init__(name,hp,mana)
        self.name=name
        self.hp = hp
        self.mana = mana
        
    def defend(self,obj):
        print("I'm the centaur, I create chaos and rule over the land")
        

print("--------------\n\n")

h = Human("Invisible Shield","Centaur","Warrior",1000,200)
m = Monster("Centaur",10000,200)

print("--------------\n\n")
h.attack(m)
print("--------------\n\n")
m.defend(h)
print("--------------\n\n")
h.attack(m)
print("--------------\n\n")

h = Human("Sword","Centaur","Warrior",1000,100)
m = Monster("Centaur",10000,200)

print("--------------\n\n")
h.attack(m)
print("--------------\n\n")
m.defend(h)
print("--------------\n\n")
h.attack(m)
print("--------------\n\n")
```
